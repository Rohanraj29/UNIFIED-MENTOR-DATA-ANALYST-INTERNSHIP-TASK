```python
data["employment_type"]=data["employment_type"].replace({"FT":"Full-Time","PT":"Part-Time","CT":"Contract","FL":"Freelance"})
```

```python
data["experience_level"]=data["experience_level"].replace({"SE":"Senior-Level","MI":"Mid-Level","EN":"Entry-Level","EX":"Executive-Level"})
```

```python
# STEP 4:- DATA VISUALIZATION
# NOW WE CAN VISUALIZE DATA TO FIND THE KEY INSIGHTS.
```

```python
a=sns.countplot(x="experience_level",data=data)
for i in a.containers:
   a.bar_label(i)
plt.title("Exprience level of Data Science Job")
plt.show()
```

# From the given data we see that most of the people are at senior level position in the company

```python
plt.figure(figsize=(20,20))
b=sns.countplot(x="job_title",data=data)
for i in b.containers:
    b.bar_label(i)
    plt.xticks(rotation=90)
    plt.xlabel("Job Title",fontsize=25)
    plt.ylabel("Count",fontsize=25)
    plt.title("Count of People by Job Title",fontsize=24)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
b=sns.countplot(x="employment_type",data=data)
for i in b.containers:
    b.bar_label(i)
    plt.xticks(rotation=90)
    plt.xlabel("Employment Type",fontsize=18)
    plt.ylabel("Count",fontsize=18)
    plt.title("Employment Type of Data Science Jobs",fontsize=18)
    plt.show()
```

# There are 588 full-time positions, This indicates that data science roles are predominantly full-time, suggesting strong demand for permanent roles in this field.

```python
et=data.groupby("employment_type")["salary"].sum().sort_values(ascending=False)
et
i=et.index
v=et.values
```

```python
plt.bar(i,v,color=["orange","blue","green","red"])
plt.xlabel("Employment Type")
plt.ylabel("Salary")
plt.title("Total Salary By Employment Type")
plt.show()
```

# The total salary for full-time roles is vastly higher than all other types.

```python
ex=data.groupby("experience_level")["salary"].sum().sort_values(ascending=False)
ex
index=ex.index
values=ex.values
```

```python
plt.bar(index,values,color=["orange","blue","green","red"])
plt.xlabel("Experience Level")
plt.ylabel("Salary")
plt.title("Total Salary By Exprience Level")
plt.show()
```

# The given data shows that the person who have a mid level exprience has a higher salary as compared to senior or entry level

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv(r"C:\Users\rohan\Downloads\Data Science Job Salaries.csv")
data.sample()
```

```python
job=data.groupby("job_title")["salary"].sum().sort_values(ascending=False).head(5)
t=job.index
d=job.values
```

```python
plt.figure(figsize=(8,8))
plt.bar(t,d,color=["green","red","orange","yellow","blue"])
plt.xlabel("Job Title")
plt.ylabel("Salary")
plt.title("Top 5 Job By Salaries")
plt.xticks(rotation=90)
plt.show()
```
